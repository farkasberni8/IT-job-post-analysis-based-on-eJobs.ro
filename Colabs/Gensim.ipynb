# Job title frequency

```python
title_list=df['title_en'].to_list()
```

```python
title_list
```

```python
!pip install --upgrade gensim
```

```python
import pprint
```

```python
from collections import defaultdict
```

```python
frequency_title= defaultdict(int)  
```

```python
frequency_title
```

```python
for text in title_list:
      frequency_title[text] += 1
```

```python
for text in frequency_title:
  if frequency_title[text] > 50:
      print(text, frequency_title[text])
```

```python
sorted_title=sorted(frequency_title.items(), key=lambda x:x[1], reverse= True)
```

```python
sorted_title
```

```python
df_new = pd.DataFrame(sorted_title, columns=['title','nr_posts'])
```

```python
df_new=df_new.head(25)
```

```python
!pip install xlsxwriter
df_new.to_excel('job_title_frequency.xlsx', engine='xlsxwriter')
```

# Company frequency

```python
company_list=df['company'].to_list()
```

```python
company_list
```

```python
frequency_company= defaultdict(int)  
```

```python
for text in company_list:
      frequency_company[text] += 1
```

```python
for text in frequency_company:
  if frequency_company[text] > 10:
      print(text, frequency_company[text])
```

```python
sorted_company=sorted(frequency_company.items(), key=lambda x:x[1], reverse= True)
```

```python
sorted_company
```

```python
df2_new = pd.DataFrame(sorted_company, columns=['company','nr_posts'])
```

```python
df2_new=df2_new.head(10)
```

```python
!pip install xlsxwriter
df2_new.to_excel('company_name_frequency.xlsx', engine='xlsxwriter')
```
